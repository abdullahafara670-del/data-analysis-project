# LOADING DATASET

```python
import pandas as pd

df = pd.read_csv("C:/Users/hp/Desktop/INTERN TASK/iris.csv")
df.head()
```

# UNDERSTANDING AND CLEANINING OF DATASET

```python
# check missing values
df.isnull().sum()

# remove duplicates
df = df.drop_duplicates()

# check data types
df.info()

# if needed, convert types (example)
# df['column_name'] = df['column_name'].astype(float)
print(df.duplicated().sum())
```

# SIMPLE STAT ON DATASET

```python
# CONDUCTING SIMPE STAT ON DATASET
df.describe()
```

# VISUALIZATION OF DATA DISTRIBUTION

```python
import matplotlib.pyplot as plt
import seaborn as sns

# histogram
df.hist()
plt.show()

# boxplot
sns.boxplot(data=df)
plt.show()

# correlation heatmap
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True)
plt.show()
```

# Basic Data Visualization And Comparism of Dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure()

sns.scatterplot(
    x='sepal_length', 
    y='sepal_width', 
    data=df,
    hue='species', # color by species
    palette='Set1'
)

plt.title("Sepal Length vs Sepal Width (by Species)")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.savefig("C:/Users/hp/Desktop/INTERN TASK/Scatter_plot.png")
plt.show()
```

```python
plt.figure()

df.plot(
    kind='line',
    title="Line Plot of Iris Features"
)

plt.xlabel("Index")
plt.ylabel("Values")

plt.savefig("C:/Users/hp/Desktop/INTERN TASK/Line_plot.png")
plt.show()
```

```python
plt.figure()

df['species'].value_counts().plot(
    kind='bar',
    color=['blue', 'green', 'red']
)

plt.title("Count of Each Species")
plt.xlabel("Species")
plt.ylabel("Count")

plt.savefig("C:/Users/hp/Desktop/INTERN TASK/Bar_plot.png")
plt.show()
```
